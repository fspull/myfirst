```python
# crop with overlap for image list
def crop_with_overlap_2(images, size, overlap_ratio):
    height,width,_ = images[0].shape
    crop_size = size
    stride = int(size*(1-overlap_ratio))
    crops = []
    
    for i in range(len(images)):
        for h in range(0,height-stride+1, stride):
            for w in range(0,width-stride+1, stride):
                if h + crop_size > height and w + crop_size > width:
                    crop = images[i][-crop_size:, -crop_size:]
                elif h + crop_size > height:
                    crop = images[i][-crop_size:, w:w + crop_size]
                elif w + crop_size > width:
                    crop = images[i][h:h + crop_size, -crop_size:]
                else:
                    crop = images[i][h:h + crop_size, w:w + crop_size]    
                crops.append(crop)
    crops = np.array(crops)
    
    return crops

def crop_with_overlap(image, size=512, overlap_ratio=0.2):
    height, width, _ = image.shape
    crop_size = size
    stride = int(size * (1 - overlap_ratio))
    crops = []

    for h in range(0, height-stride+1, stride):
        for w in range(0, width-stride+1, stride):
            if h + crop_size > height and w + crop_size > width:
                crop = image[-crop_size:, -crop_size:]
            elif h + crop_size > height:
                crop = image[-crop_size:, w:w + crop_size]
            elif w + crop_size > width:
                crop = image[h:h + crop_size, -crop_size:]
            else:
                crop = image[h:h + crop_size, w:w + crop_size]
            crops.append(crop)

    crops = np.array(crops)

    return crops
```
